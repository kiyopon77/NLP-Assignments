importing libraries and downloading important resources


```python
# importing libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer

# download required nltk resources
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

Defining the paragraph.

```python
paragraph = (
    """Cinema, once a sanctuary for artistic expression and cultural critique, has steadily degenerated under the weight of venture capitalist influence. Profit-driven conglomerates 
    now dominate the medium, reducing filmmaking to algorithmic spectacle rather than creative exploration. The rise of franchises like Marvel epitomizes this shift, films engineered 
    for mass consumption, brand synergy, and endless sequels rather than aesthetic or emotional depth. The artistry that once defined cinema has been supplanted by market-tested formulas 
    and visual overload. What was once an art form reflecting humanity's complexities has become a product line, optimized for investors, not audiences hungry for truth or beauty."""
)
print(paragraph)
```

Tokenization : This splits the paragraph into individual words and punctuation marks


```python
tokens = word_tokenize(paragraph)
print("Token")
print(tokens)
```

Stopword Removal : This removes common english words like "is","the","and",etc. Keeping only meaningful words.

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]
print("after stopword removal")
print(filtered_tokens)
```

Stemming : Reduce a word to its root forms often not actual words

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("after stemming")
print(stemmed_words)
```

Lemmatization : lemmatization returns dictionary base forms.

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("After lemmatization")
print(lemmatized_words)
```
